## One fact

```python
# read image through opencv api
img_bgr = cv2.imread('../data/puppy.jpg')

# display using matplotlib
plt.imshow(img_bgr)
```

Observation :
- OpenCV read image as a numpy array, but the color of each pixel has been sorted as <span class="girk">BGR</span>.
- Matplotlib's plot expects an <span class="girk">RGB</span> image.
- It is necessary to swap those channels.
- This operation can be done either by using openCV conversion functions <span class="mark">cv2.cvtColor()</span> or by working directly with the numpy array.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.imread('../data/puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```python
img_gray = cv2.imread('../DATA/00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray,cmap='gray')
img_gray.shape
```

## Resize Images

```python
img_rgb.shape
# width, height, color channels
```

```python
img =cv2.resize(img_rgb,(1300,275))
```

```python
plt.imshow(img)
```

### By ratio

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img =cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
%matplotlib qt
plt.imshow(new_img)
```

### Flipping Images

```python
%matplotlib inline
```

```python
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```
